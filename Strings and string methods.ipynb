## String methods
- Case Conversion Methods
  - upper():It converts entire string into upper case
  - lower():It converts entire string into lower case
  - title():It converts the first letter of every word into upper and remaining lower
  - swapcase():It converts upper case letters into lower case and lower case into upper case
  - capitalize():It converts first character into upper case and remaining into lower case.

```python
string="The Sky is blue"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

## Remove white space or pattern at ends of string
  -strip():By default it removes the white space at both ends of string
  -lstrip():It removes the white space at left side of string
  -rstrip():It removes the white space at right side of string

```python
s="     Sreedevi     "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@@Sree@@@devi@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip())
```

```python
#split():It splits the entire string into list of substring based on specified seperator
#by default it splits at space
string="The sky is blue and ocean is blue"
print(string.split())
print(string.split("is"))
print(string.split("e"))
```

```python
#'sep'.join(iterable):It joins the all iterable elements with seperator.
w=['The', 'Sky', 'Blue', 'and', 'Ocean', 'is', 'Blue']
print("".join(w))
print(" ".join(w))
print("-".join(w))
```

# Substring index finding
 -index(substring):if substrings is exists in string,it returns the first occurence of substring position otherwise 'Value Error'
 -find(substring):if substrings is exists in string,it returns the firt occurence of substring position,otherwise -1

```python
string="The sky is blue and ocean is blue"
print(string.index('is'))
print(string.index('z'))
```

```python
string="The sky is blue and ocean is blue"
print(string.find('is'))
print(string.find('z'))
print(string.find('is', 10))
print(string.find('e', 10))
print(string.find('e', -5))
print(string.find('v', 3))    #if character not there in string it return -1
```

```python
#count(substring):It returns the total non-overlapping occurences of sustring
s="ABCBCBababaabb"
print(s.count("B"))
print(s.count("BCB"))
lower_string=s.lower()
print(lower_string)
print(lower_string.count('b'))
print(s.upper().count('B'))
print(s.count('z'))
```

```python
#replace(old_string,new_string)=It replaces the old sub string with new substring in string
string="The sky is blue and ocean is Blue"
print(string.replace("blue", "Red"))     #here small b only replace
print(string.replace("e", "E"))
print(string.replace("e", "E", 2))   #Replaces the first 2 occurences
```

## String Checking Methods
*String checking methods returns the boolean value as result*
-isupper():It return True, If entire string is in upper case
-islower():It return True, If entire string is in lower case
-isalpha():It return True, If entire string contains alphabets
-isdigit():It return True, If entire string contains digits
-isalnum():It return True, If entire string contains either alphabets or digits
-startswith(substring):It return True, If string starts with substring
-endswith(substring):It return True, If string ends with substring

```python
print("CODE".isupper())
print("CODE".islower())
print('code'.islower())
print('123'.isdigit())
print('123abc'.isdigit())
print('Code'.isalpha())
print('Code123'.isalnum())
print('123'.isalnum())
print('Code'.startswith('c'))
print('Code'.endswith('e'))
```
