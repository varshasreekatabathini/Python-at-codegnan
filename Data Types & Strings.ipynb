```python
#indexing
s="Mani"
print(s[2])
print(s[3])
```

```python
#slicing
#str_obj[start:end:step]
s="Indra"
print(s[0:4:1])
```

```python
s="Codegnan"
print(s[2:7])
print(s[-4:])
print(s[5:2])
print(s[3:100])
print(s[::-1])
print(s[:5:])
print(s[-3::-1])
print(s[-3::1])
```

```python
#empty string representation
#type() : This function returns the objects datatype
s1=""
s2=str()
print(type(s1))
print(type(s2))
```

```python
#string reading from user
#input() : This function is used to read dynamically from user and this function by default read it as string data type
s=input()
print(type(s))
print(s)
```
